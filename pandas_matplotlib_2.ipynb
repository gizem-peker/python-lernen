```python
age_order = [
    '<18 years old',
    '18-24 years old',
    '25-34 years old',
    '35-44 years old',
    '45-54 years old',
    '55-64 years old',
    '>65 years old'
]
#zählt alle werte
age_counts = df['age'].value_counts()
#bringt sie in richtige Reihenfolge
age_counts = age_counts.reindex(age_order)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.bar(age_counts.index, age_counts.values, color='coral')

plt.title('Altersverteilung der Teilnehmer')
plt.xlabel('Altersgruppe')
plt.ylabel('Anzahl Teilnehmer')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# bar Plot
plt.figure(figsize=(10, 6))
plt.bar(age_counts.index, age_counts.values, color='steelblue')

plt.title('Altersverteilung der Teilnehmer')
plt.xlabel('Altersgruppe')
plt.ylabel('Anzahl Teilnehmer')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
pref_counts = df['prefer_overall'].value_counts()
```

```python
#pie Plot

plt.figure(figsize=(7, 7))
plt.pie(pref_counts.values,labels=pref_counts.index,autopct='%1.1f%%', colors=['steelblue', 'coral', 'mediumseagreen', 'mediumpurple'],startangle=90)
#color=[...]Verschiedene Farben für jeden Balken
#autopct='%1.1f%%'Zeigt Prozent im Pie Chart an
#startangle=90 Pie Chart startet oben statt rechts

plt.title('Welchen Kaffee bevorzugen die Teilnehmer?')

plt.tight_layout()
plt.show()
```
